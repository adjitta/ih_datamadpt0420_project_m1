SCRAPING WEB

```python
country_code_url = 'https://ec.europa.eu/eurostat/statistics-explained/index.php/Glossary:Country_codes'
```

```python
country_code_page = requests.get(country_code_url)
soup_country_code = BeautifulSoup(country_code_page.content,'lxml')
```

```python
country_table = soup_country_code.find_all('table')
```

```python
country_table_text = []
```

```python
for country in country_table:
    country_table_text.append(country.text)
```

```python
country_table_text
```

```python
country_table_split = list(map(lambda x: x.split('\n'), country_table_text))
country_table_split
```

```python
country_split = []

for country in country_table_split:
    for country_code in country:
        if (country_code != '') and (country_code != '\xa0') and (country_code != ' \xa0'):
            country_split.append(country_code)
```

```python
country_split
```

```python
country_split[0:2]
```

```python
row_split = 2

rows_refactored = [country_split[x:x+row_split] for x in range(0, len(country_split), row_split)]

rows_refactored
```

```python
country_code_df = pd.DataFrame(rows_refactored, columns=['country', 'country_code'])
```

```python
country_code_df
```

```python
country_code_df['country_code']= country_code_df['country_code'].str.replace('(', '')
country_code_df['country_code']= country_code_df['country_code'].str.replace(')', '')
```

```python
country_code_df.replace(to_replace=["EL"],value=["GR"],inplace=True)
country_code_df['country_code']= country_code_df['country_code'].str.replace('UK', 'GB')
country_code_df.head(30)
```

JOIN TABLES

```python
country_code_merge = country_info.merge(country_code_df,  how = 'inner', left_on = 'country_code', right_on = 'country_code')
```

```python
country_code_merge
```

```python
career_merge = country_code_merge.merge(career_info, how = 'left', left_on = 'uuid', right_on = 'uuid')
```

```python
data_merge = career_merge.merge(jobs_tittle, how = 'left', left_on = 'normalized_job_code', right_on = 'uuid')    
```

```python
data_merge.head()
```

```python
final_data_columns = data_merge[['uuid_x', 'country', 'dem_education_level','dem_full_time_job', 'normalized_job_title', 'rural']]
```

```python
final_data_columns
```

```python
final_data_columns['rural'] = final_data_columns['rural'].str.replace('city','urban')
final_data_columns['rural'] = final_data_columns['rural'].str.replace('Non-Rural','urban')
final_data_columns['rural'] = final_data_columns['rural'].str.replace('Country','rural')
final_data_columns['rural'] = final_data_columns['rural'].str.replace('countryside','rural')
```

```python
final_data_columns
```

```python
final_data_columns.rename(columns={'uuid_x':'ID',
                                   'dem_education_level':'Education level',
                                   'dem_full_time_job':'Full time job',
                                   'normalized_job_title':'Job Title',
                                   'rural':'lifestyle'},
                          inplace=True)
```

```python
grouped = final_data_columns.groupby(['country', 'Job Title', 'lifestyle']).agg({'ID': 'count'})
grouped.columns = ['Quantity']
grouped = grouped.reset_index()
grouped['Percentage'] = (grouped['Quantity'] / grouped['Quantity'].sum()) * 100
grouped['Percentage'] = grouped['Percentage'].apply(lambda x: "{0:.2f}%".format(x))
grouped['Job Title'] = grouped['Job Title'].apply(lambda x: x.capitalize())
grouped.sort_values(by='Quantity', ascending=False)
```

```python
grouped.to_csv('./data/results/People_jobs_info.csv', header=False, index=False)
```

```python
def group_data(data):
    grouped = data.groupby(['country', 'Job Title', 'lifestyle']).agg({'ID': 'count'})
    grouped.columns = ['Quantity']
    grouped = grouped.reset_index()
    grouped['Percentage'] = (grouped['Quantity'] / grouped['Quantity'].sum()) * 100
    grouped['Percentage'] = grouped['Percentage'].apply(lambda x: "{0:.2f}%".format(x))
    grouped['Job Title'] = grouped['Job Title'].apply(lambda x: x.capitalize())
    grouped.sort_values(by='Quantity', ascending=False)
    return grouped
```
